```python

num_months = 12
date_range = pd.date_range(start=pd.Timestamp.now() - pd.DateOffset(months=num_months), periods=num_months, freq='M')

savings_data = pd.DataFrame({
    'ds': date_range,
    'y': np.random.rand(num_months) * 1000  
})


model = Prophet()
model.fit(savings_data)


future = model.make_future_dataframe(periods=12, freq='M')  # Forecast for the next 12 months

forecast = model.predict(future)

fig = model.plot(forecast)
plt.title("Forecast of Mutual Funds Growth")
plt.xlabel("Date")
plt.ylabel("Mutual Funds Amount")
plt.show()

```
